```python
import pandas as pd
df = pd.read_csv('FIFA.csv',encoding='euc-kr')
df.head()
```

1. FIFA 데이터에 각 선수의 키는 Height변수에 피트와 인치로 입력되어 있다. 이를 cm로 변환하여 새로운 변수 Heihht_cm을 생성하시오. 

("'" 앞의 숫자는 피트이며 "'"뒤의 숫자는 인치, 1피트 = 30cm, 1인치 = 2.5cm)

```python
df['Height_cm'] = df.Height.str.split("'").str[0].astype(int)*30+df.Height.str.split("'").str[1].astype(int)*2.5 
df
```

2. 포지션을 의미하는 포지션 변수를 아래 표를 참고하여 'Forward', 'Midfielder', 'Defender' 'GoalKeeper'로 재범주화하고 factor형으로 변환하여 Position_Class라는 변수르르 생성하고 저장하시오.

```python
df['Position_Class'] = ""
Forward = ['LS','ST','RS','LW','CF','RF','RW']
Midfielder = ['LAM','CAM','RAM','LM','LCM','CM','RCM','RM']
GoalKeeper = ['GK']
Defender = ['LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']

for i in range(len(df)):
    if df['Position'][i] in Forward:
        df['Position_Class'][i] = 'Forward'
        
    elif df['Position'][i] in Midfielder:
        df['Position_Class'][i] = 'Midfielder'
        
    elif df['Position'][i] in Defender:
        df['Position_Class'][i] = 'Defender'
        
    elif df['Position'][i] in GoalKeeper:
        df['Position_Class'][i] = 'GoalKeeper'
        
df.head()
```

3. 새로 생성한 Position_Class 변수의 각 범주에 따른 Value 변수 평균값의 차이를 비교하는 일원배치 분산분석을 수행하고 결과를 해석하시오. (데이터는 등분산성 가정을 만족한다고 가정) 그리고 평균값의 차이가 통계적으로 유의하다면 사후검정을 수행하고 설명하시오.

H0 : k개의 집단 간 모평균에는 차이가 없다. 

H1 : k개의 집단 간 모평균이 모두 같다고는 할 수 없다. 

```python
df['Position_Class'].value_counts()
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.stats.anova import AnovaRM
from scipy import stats

anova = smf.ols(formula = 'Value~Position_Class',data = df)
anova_fit = anova.fit()

anova_table = sm.stats.anova_lm(anova_fit,typ=2)
print(anova_table)
```

p-value값이 0.05보다 작아 귀무가설 기각 -> 사후검정 필요

```python
from statsmodels.stats.multicomp import MultiComparison
from statsmodels.stats.multicomp import pairwise_tukeyhsd
hsd = pairwise_tukeyhsd(df['Value'], df['Position_Class'], alpha=0.05)
print(hsd.summary())
```

Forward-MidFielder의 pvalue값이 0.5

```python
#봉페로니 교정
comp = MultiComparison(df.Value, df.Position_Class)
result = comp.allpairtest(stats.ttest_ind, method='bonf')
print(result[0])
```
