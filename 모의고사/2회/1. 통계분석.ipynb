```python
import pandas as pd
import numpy as np

data = pd.read_csv("Admission.csv")
data.head()
```

```python
corr_matrix = data.corr()
corr_matrix
```

```python
corr_matrix['Chance_of_Admit'].sort_values(ascending=False)
```
